```python
import pandas as pd
```

```python
demographics_pd = pd.read_csv("DEMOGRAPHICS.csv")
demographics_pd
```

```python
risk_factors_pd = pd.read_csv("RISKFACTORSANDACCESSTOCARE.csv")
risk_factors_pd
```

```python
vulnerable_pd = pd.read_csv("VUNERABLEPOPSANDENVHEALTH.csv")
vulnerable_pd
```

```python
merged_pd = demographics_pd.merge(vulnerable_pd).merge(risk_factors_pd)
```

```python
merged_pd
```

```python
merged_pd = merged_pd[merged_pd['Obesity'] > 0]
merged_pd = merged_pd[merged_pd['No_Exercise'] > 0]
merged_pd = merged_pd[merged_pd['Few_Fruit_Veg'] > 0]
```

```python
for i in range(3141):
    if i in merged_pd.index:
        merged_pd["No_HS_Diploma"][i] = "{:.1f}".format(
            (float(merged_pd["No_HS_Diploma"][i]/merged_pd["Population_Size"][i]) * 1000)/10)
    else:
        pass
```

```python
for i in range(3141):
    if i in merged_pd.index:
        merged_pd["Major_Depression"][i] = "{:.1f}".format(
            (float(merged_pd["Major_Depression"][i]/merged_pd["Population_Size"][i]) * 1000)/10)
    else:
        pass
```

```python
for i in range(3141):
    if i in merged_pd.index:
        merged_pd["Non-White"][i] = 100 - merged_pd["White"][i]
    else:
        pass
```

```python
for i in range(3141):
    if i in merged_pd.index:
        if float(merged_pd["Uninsured"][i]/merged_pd["Population_Size"][i]) * 100 > 0:
            merged_pd["Uninsured"][i] = "{:.1f}".format(
                (float(merged_pd["Uninsured"][i]/merged_pd["Population_Size"][i]) * 1000)/10)
        else:
            merged_pd["Uninsured"][i] = 0.0
    else:
        pass
```
