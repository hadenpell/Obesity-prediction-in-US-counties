# Important info about the dataset:
* Question / topic = Weight classification by BMI
* Response = The category they were measuring (percentage of obese, underweight, overweight, or normal)
* Break out category = What class they're measuring (race, income, etc)
* Break out = the specific group in the category measured 
* Data value:
The percentage is a weighted percentage for that response.
The weighted percentage is adjusted to represent the entire state.
So 24.9% of the state population are calculated as being obese.

ResponseID:
039 = Obese
040 = Overweight
041 = Normal Weight
042 = Underweight

IncomeID: (BreakoutID)
INCOME1 - < $15,000
INCOME2 - $15,000 - $24,999
INCOME3 - $25,000 - $34,999
INCOME4 - $35,000 - $49,999
INCOME5 - $50,000+

Data value unit: Percentage

```python
obesity_df = pd.read_csv('Obesity_Data.csv')
obesity_df
```

```python
#clean data initially

cols_needed = ["Year","Locationabbr", "Response", "Break_Out", "Break_Out_Category", "Sample_Size", "Data_value",
              "BreakoutID", "ResponseID", "GeoLocation"]

clean_cols = obesity_df[cols_needed]

clean_cols
```

```python
income_only = clean_cols[clean_cols["Break_Out_Category"] == "Education Attained"]

# race_and_income = clean_cols[(clean_cols["Break_Out_Category"] == "Household Income") | (clean_cols["Break_Out_Category"]
# == "Race/Ethnicity")]

# income_1_only = clean_cols[(clean_cols["Break_Out_Category"] == "Household Income") & (clean_cols["BreakoutID"]
# == "INCOME1") | (clean_cols["BreakoutID"]
# == "INCOME2")]

income_only
```

```python
import seaborn as sns
```

```python
#Obesity prevalence by income status
obese_by_income = income_only[income_only["ResponseID"]=="RESP039"]

#Percentage of obese people by income

graph = sns.barplot(x="BreakoutID", y="Data_value", color="red", data=obese_by_income, 
                        order=["INCOME1", "INCOME2", "INCOME3",
                  "INCOME4", "INCOME5"])

#add labels to the axes  
graph.set_xlabel("Income Status", size = 16)
graph.set_ylabel("Obesity Prevalence, %", size = 16)
```

```python
#function to replace income breakoutIds with numbers
def clean_income(income):
    if income=="INCOME1":
        return "<$15,000"
    elif income=="INCOME2":
        return "$15,000 - $24,999"
    elif income=="INCOME3":  
        return "$25,000 - $34,999"
    elif income=="INCOME4":
        return "$35,000 - $49,999"
    elif income=="INCOME5":
        return "$50,000"

#create data frame where income ids are replaced by their corresponding numerical values
income_ids_num = obese_by_income.replace("INCOME1", "<$15,000")
income_ids_num = income_ids_num.replace("INCOME2", "$15,000 - $24,999")
income_ids_num = income_ids_num.replace("INCOME3", "$25,000 - $34,999")
income_ids_num = income_ids_num.replace("INCOME4", "$35,000 - $49,999")
income_ids_num = income_ids_num.replace("INCOME5", "$50,000")

graph1 = sns.scatterplot(x="BreakoutID", y="Data_value", color="red", data=income_ids_num) 
                      #  order=["<$15,000", "$15,000 - $24,999", "$25,000 - $34,999",
                  #"$35,000 - $49,999", "$50,000"])
```

```python
#The percentage of obesity by year 
obese_by_income.groupby("Year").mean()
```

```python
#prevalence rates of obesity by income, 2011-2019

obesity_by_income = income_ids_num.groupby("BreakoutID")
obesity_by_income
```
